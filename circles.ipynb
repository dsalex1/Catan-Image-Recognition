```python
#load all images from the circle folder and display them as plot

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import pytesseract


#load all images from the circle folder and display them as plot
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        #print(img)
        if img is not None:
            images.append(img)
    return images

images = load_images_from_folder("saved")
#load and print all the timages
_, axs = plt.subplots(8, 15, figsize=(15, 15))
axs = axs.flatten()
for img, ax in zip(images, axs):
    ax.imshow(img)
    text = pytesseract.image_to_string(
        img, config='--psm 9  -c tessedit_char_whitelist=12345689')
    ax.set_title(''.join(filter(str.isdigit, text)))
    ax.axis('off')
plt.show()

```
